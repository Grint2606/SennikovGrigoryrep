```python
def mult(a, b):
    if len(a[0]) != len(b):
        raise ValueError("Неправильные размеры матриц для умножения")
    result = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]
    result = [
    [sum(a[i][k] * b[k][j] for k in range(len(b))) for j in range(len(b[0]))]
    for i in range(len(a))
]

    return result
```

```python
import numpy as np

def np_mult(a, b):
    result = np.dot(a, b)
    return result
```

без numpy

```python
def scalar_product(a, b):
    if len(a) != len(b):
        raise ValueError("Длины векторы должны быть одинаковы")

    s = 0
    for x, y in zip(a, b):
        s += x * y
    return s
```

с numpy

```python
def np_scalar_product(a, b):
    s = np.dot(a, b)
    return s
```

```python
%time product_1 = scalar_product(a,b)
%time product_2 = np_scalar_product(a,b)
# проверим корректность:
np.abs(product_1 - product_2).sum()
```

Задание 3 Напишите функцию, которая находим сумму четных диагональныех элементов квадратной матрицы. Если таких элементов нет, то вывести 0.

```python
# функция, решающая задачу с помощью NumPy
def np_diag_2k(a):
    diagonal_elements = np.diagonal(a)
    even_diagonal_elements = diagonal_elements[diagonal_elements % 2 == 0]
    res = even_diagonal_elements.sum() if even_diagonal_elements.size > 0 else 0
    return res
```

```python
%%time
# засечем время работы функции с NumPy
np_diag_2k(a)
```

Задание 4
На вход дан двумерный массив  X . Напишите функцию, которая для каждой строчки  x=(x1,x2,…,xn)  массива  X  строит строчку  s=(s1,s2,…,sn) , где  sk=x1+...+xkk , а затем выдаёт массив из построенных строчек. Используйте библиотеку numpy (вам поможет функция cumsum). Выходом функции должен быть двумерный массив той же формы, что и  X .

```python

import scipy.stats as sps

def np_sec_av(*A):
    stacked_arrays = np.vstack(A)
    sorted_rows = np.partition(stacked_arrays, 1, axis=1)
    second_largest_values = sorted_rows[:, 1]
    return second_largest_values
```

```python
# зададим некоторую последовательность и проверим ее на ваших функциях.
# функция должна работать ~ в 50 раз быстрее, чем без Numpy
A = sps.uniform.rvs(size=10**3)
print(A)

%time S2 = np_sec_av(A)
```

Задание 5
Написать функцию для кодирование массива (Run-length encoding). Все подряд повторения элементов функция сжимает в один элемент и считает количество повторений. Функция возвращает кортеж из двух векторов одинаковой длины. Первый содержит элементы, а второй — сколько раз их нужно повторить.

Пример: encode(np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])) = (np.array[1, 2, 3, 1, 5, 2, 3]), np.array[1, 2, 2, 2, 2, 1, 2])

```python
# функция, решающая задачу с помощью NumPy
def np_encode(a):
    if len(a) == 0:
        return np.array([]), np.array([])

    # Индексы, где происходит смена элементов
    change_indices = np.where(a[:-1] != a[1:])[0]

    # Добавляем индекс конечного элемента
    change_indices = np.concatenate(([0], change_indices + 1, [len(a)]))

    # Значения и их количество
    encoded_values = a[change_indices[:-1]]
    counts = np.diff(change_indices)

    result = encoded_values, counts
    return result
```

Задание 6
Пусть задан некоторый одномерный массив  X . Надо построить новый массив, где все элементы с нечетными индексами требуется заменить на число  a  (если оно не указано, то на 1). Все элементы с четными индексами нужно возвести в куб. Затем записать элементы в обратном порядке относительно их позиций. Массив  X  при этом должен остаться без изменений. В конце требуется слить массив X с преобразованным X и вывести в обратном порядке.

```python
# функция, решающая задачу с помощью NumPy
import numpy as np

def np_transformation(X, a=1):
    result = np.copy(X)
    result[1::2] = a
    result[::2] **= 3
    result = np.concatenate((X, result[::-1]))
    return result


```

```python
X = np.array([i for i in range(1, 10, 2)])
# функция должна работать ~ в 20 раз быстрее, чем без Numpy
%time S2 = np_transformation(X, 5)
```
